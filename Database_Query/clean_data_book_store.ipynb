```python
book_clean_x = book.drop_duplicates()
book_clean_x.duplicated().any()
```

```python
book_clean_x.sort_values(by='price')
```

```python
book_clean_y = book_clean_x[book_clean_x['price'] != 1]
```

```python
import numpy as np
```

```python
book_clean_y.loc[:, 'score'] = np.random.randint(0, 6, size=len(book_clean_y))
book_clean_y
```

```python
book_clean_y.to_csv('book_data.csv', index=False, encoding='utf-8-sig')
```

```python
book_clean_y.to_sql('book', con=engine, if_exists='append', index=False)
```
